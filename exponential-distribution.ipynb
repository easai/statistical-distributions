# 指数分布

指数分布は確率密度関数(PDF)が指数である確率分布です。

$$
f(x)=\lambda e^{-\lambda x}
$$

$x$ のとれる範囲は $[0,\infty]$ で、積分すれば１となります。

$$
\int_0^\infty \lambda e^{-\lambda x}=\left. -e^{-\lambda x}\right|_0^\infty=1
$$

累積分布関数(CDF)は確率密度関数を積分し、$x=\infty$ で $F(x)=１$ なので積分定数が計算できて、

$$
F(x)=1-e^{-\lambda x}
$$



```python
import matplotlib.pyplot as plt
import numpy as np


class ExpDist():
    def __init__(self, n):
        self.n = n

    def pdf(self, x):
        return self.n*np.exp(-self.n*x)

    def cdf(self, x):
        return 1-np.exp(-self.n*x)


a = np.arange(0, 4, .01)
x = np.linspace(0, 10, 100)
l = []
for i in range(1, 5):
    expdist = ExpDist(i)
    y = expdist.pdf(a)
    plt.plot(a, y)
    l.append("n = "+str(i))

plt.legend(l)
plt.xlabel("x")
plt.ylabel("f(x)")
plt.title("Exponential Distribution PDF")
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

a = np.arange(0, 6, .01)
x = np.linspace(0, 10, 100)
l = []
for i in range(1, 5):
    expdist = ExpDist(i)
    z = expdist.cdf(a)
    plt.plot(a, z)
    l.append("n = "+str(i))

plt.legend(tuple(l))
plt.xlabel("x")
plt.ylabel("F(x)")
plt.title("Exponential Distribution CDF")
plt.show()
```
